```python
import pyaudio
import wave

# function that reads wav file to array
def read_wav( wavname ):
    
    wf = wave.open(wavname, 'rb')
    
    CHUNK = 1024
    frames = []
    data_str = wf.readframes(CHUNK) #read a chunk
    
    while data_str != '':
        data_int = np.fromstring( data_str, 'int16') # convert from string to int (assumes .wav in int16)
        data_flt = data_int.astype( np.float32 ) / 32767.0 # convert from int to float32
        frames.append( data_flt )  #append to list
        data_str = wf.readframes(CHUNK) #read a chunk

    return np.concatenate( frames )


def play_audio( data, p, fs):
    # data - audio data array
    # p    - pyAudio object
    # fs    - sampling rate
    
    # open output stream
    ostream = p.open(format=pyaudio.paFloat32, channels=1, rate=fs,output=True)
    # play audio
    ostream.write( data.astype(np.float32).tostring() )
```

```python
## Play sounds
#fname = "gqrx-20140508-070156.wav"
fname = "prefix_1.wav"
fs = 48000

# get sound file
data = read_wav( fname )
```

```python
# instantiate PyAudio
p = pyaudio.PyAudio()
# play sound
play_audio( data, p, fs )
    
# terminate pyAudio
p.terminate()
```

```python
fs = 48000
def demod_and_plot(QAM, title_string):
    f = np.linspace(-fs/2,fs/2, len(QAM))
    t = np.r_[0.0:len(QAM)*1.0]/fs
    fig = figure(figsize = (16,4))
    plot(QAM)
    title('original %s'%(title_string))
    fig = figure(figsize = (16,4))
    plot(f, abs(fft.fftshift(fft.fft((QAM)))))
    title('original fft %s'%(title_string))
    
    f0=1800
    
    #real
    r,i = demod_QAM16(QAM, t, f0=900, fs=48000)
    fig = figure(figsize = (16,4))
    plot(r)
    title('real %s'%(title_string))
    fig = figure(figsize = (16,4))
    plot(f, abs(fft.fftshift(fft.fft((r)))))
    title('real spec %s'%(title_string))
    
    #imag
    fig = figure(figsize = (16,4))
    title('imag %s'%(title_string))
    plot(i)
    fig = figure(figsize = (16,4))
    plot(f, abs(fft.fftshift(fft.fft((i)))))
    title('imag spec %s'%(title_string))
```
