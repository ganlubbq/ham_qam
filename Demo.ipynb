```python
code = np.array((-2-2j, -2-1j,-2+2j,-2+1j,-1-2j,-1-1j,-1+2j,-1+1j,+2-2j,+2-1j,+2+2j+2+1j,1-2j,+1-1j,1+2j,1+1j))/2

np.random.seed(seed=1)
rbits = np.int16(rand(6,1)*15)
prefix = np.array([[0],[2],[10],[8]])#, [0],[2],[10],[8]])
bits = np.int16(rand(26,1)*15)

Nbits = len(rbits) + len(prefix) + len(bits)  # number of bits
print "Generating %d random bits"%(Nbits)
bits = np.array(rbits.tolist() + prefix.tolist() + bits.tolist())
```

Modulate

```python
QAM, M, t, M_prefix, t_prefix = mod_QAM16(bits, prefix, f0=1800, tbw=4, fs=48000, baud=300, shaped=False, plot=False)
```

Demodulate

```python
r,i = demod_QAM16(QAM, t, f0=1800, fs=48000)
```

Detect Sync

```python
corr_index = detect_sync(r, i, M_prefix, len(prefix))
```

Decode

```python
symbols = decode_symbols(r, i, corr_index, M.real.ravel(), M.imag.ravel(), Nbits-len(rbits))
```
