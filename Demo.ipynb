###Transmitter
####Modulation
Using a 16 symbol QAM scheme allows a simple mapping of hex values to particular symbols. We first append a known sequency of 8 values corresponding to the corners of the QAM 16 constellation in order to detect the beginning of a transmission. In order to create the QAM signal, we first map the symbols to the complex values they represent. We next split the signal into it's real and imaginary parts (in phase denoted as $I$ and quadrature denoted as $Q$). To generate the QAM signal by multiplying $I$ by $cos(2 \pi f_0 t)$ and $Q$ by $-sin(2 \pi f_0 t)$. We chose a carrier frequency of 1800 Hz. We then output the sum of these components as audio which was FM modulated and transmitted by the handheld radio.

###Receiver
####Demodulating
In order to demodulate the output we first multiplied the $I$ and $Q$ components by $cos(2\pi f_0)$ and $-sin(2\pi f_0)$ respectively.  Next we applied an FIR low pass filter to remove the carrier frequency.

####Syncing
In order to detect the beginning of a transmission we inserted a known prefix of 8 values as previously explained. These values were chosen specifically to be the corners of the constellation as in addition to being $90^{\circ}$ out of phase with each other will also have the maximum amplitude. Because QAM encodes data using both the amplitude and phase of a signal, we can use the maximum amplitude of the input signal in order to normalize the data before decoding. To detect the index at which our transmission begins we identify the maximum of the correlation between the received signal and the known prefix. This method can easily be extended to a packet protocol with a known packet header. 

####Decoding
To decode the demodulated signal we detect the values of the $I$ and $Q$ components separately. We first normalize each signal based on the maximum amplitude of the detected prefix, and then scale to a range of -2 to 2. This scaling provides the convenience of being able to easily quantize the signal in the range of values corresponding to the constellatoin. We then round the data and simply detect the value of the signal in the middle of each bit period for a given number of bits, as this should be the index of maximum correlation if the data is properly synced. Because our original goal was to use this system for sending packets, we designed it to decode a known number of symbols.

###Generating simulation data
In order to simulate the system, we generated a random sequence of symbols. We also generate a random sequence of symbols to prepend to the transmission in order to demonstrate the process of syncing.

```python
np.random.seed(seed=1)
rbits = np.int16(rand(6,1)*15)
prefix = np.array([[0],[2],[10],[8], [0],[2],[10],[8]])
bits = np.int16(rand(26,1)*15)

Nbits = len(rbits) + len(prefix) + len(bits)  # number of bits
print "Generating %d random bits"%(Nbits)
bits = np.array(rbits.tolist() + prefix.tolist() + bits.tolist())
```

##Modulate
Add the sync prefix and modulate the data.

```python
QAM, M, t, M_prefix, t_prefix = mod_QAM16(bits, prefix, f0=1800, tbw=4, fs=48000, baud=300, shaped=False, plot=True)
```

##Modulate
Demodulate and separate the raw signal into its $I$ and $Q$ components.

```python
i,q = demod_QAM16(QAM, t, f0=1800, fs=48000)
```

##Sync
Detect the prefix sequence to sync with the begnning of a transmission. We can note that the correlation has a number of local maxima as the random symbols at the beginning of the transmission, however in a real system this would most likely be noise when data is not being transmitted and therefore be more identifiable.

```python
corr_index = detect_sync(i, q, M_prefix, len(prefix), plot=True)
```

##Decoding
The following figures demonstrate the quantization and detection of symbols.

```python
symbols = decode_symbols(i, q, corr_index, M.real.ravel(), M.imag.ravel(), Nbits-len(rbits), plot=True)
```

For the ideal case we can confirm that all symbols were transmitted correctly and we were able to identify the symbols corresponding to our simulated transmission.

```python
print symbols
#print bits.tolist()[6:]
decoded = symbols_to_hex(symbols)
print 'Errors: %d'%(np.sum([s[0]!=s[1][0] for s in zip(decoded, bits.tolist()[6:])]))
```
